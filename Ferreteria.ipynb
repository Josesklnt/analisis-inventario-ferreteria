```
import pandas as pd
import numpy as np

# Creamos 1000 productos ficticios para tu ferretería
n = 1000
data = {
    'ID': range(1, n + 1),
    'Producto': np.random.choice(['Martillo', 'Taladro', 'Tornillo', 'Pintura', 'Llave'], n),
    'Stock': np.random.randint(0, 100, n),
    'Precio': np.random.uniform(10, 500, n).round(2)
}
df = pd.DataFrame(data)
df.to_csv('inventario.csv', index=False)
print("¡Archivo de inventario creado!")
```

```
# Vamos a ver los productos que se están agotando (menos de 5 unidades)
faltantes = df[df['Stock'] < 5]
print("--- PRODUCTOS POR AGOTARSE ---")
print(faltantes.head())

# Calculamos cuánto dinero hay invertido en total
total_dinero = (df['Stock'] * df['Precio']).sum()
print(f"\nInversión total en la ferretería: ${total_dinero:,.2f}")
```
