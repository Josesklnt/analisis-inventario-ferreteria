```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- PASO 1: GENERAR LOS DATOS ---
np.random.seed(42)
n = 1000
df = pd.DataFrame({
    'ID': range(1, n + 1),
    'Producto': np.random.choice(['Martillo', 'Taladro', 'Tornillo', 'Pintura', 'Llave'], n),
    'Stock': np.random.randint(0, 100, n),
    'Precio': np.random.uniform(10, 500, n).round(2)
})

# --- PASO 2: ANÁLISIS DE PARETO ---
df['Valor_Stock'] = df['Stock'] * df['Precio']
df_pareto = df.sort_values(by='Valor_Stock', ascending=False).reset_index(drop=True)
df_pareto['Porcentaje_Acumulado'] = 100 * df_pareto['Valor_Stock'].cumsum() / df_pareto['Valor_Stock'].sum()

# --- PASO 3: GRAFICAR ---
plt.figure(figsize=(10,5))
plt.plot(df_pareto.index, df_pareto['Porcentaje_Acumulado'], color='blue', linewidth=2)
plt.axhline(y=80, color='r', linestyle='--', label='80% del Valor (Pareto)')
plt.title('Análisis de Inversión en Inventario (Regla 80/20)')
plt.xlabel('Cantidad de Productos (Ordenados por Valor)')
plt.ylabel('% Acumulado de Inversión')
plt.legend()
plt.grid(True)
plt.show()

print(f"Inversión total: ${df['Valor_Stock'].sum():,.2f}")
```

```
# --- EXPERIENCIA 3: SIMULACIÓN DE VENTAS MENSUALES ---
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
# Simulamos ventas con un pico en Diciembre (temporada alta)
ventas = [12000, 11500, 13000, 14500, 16000, 15500, 14000, 15000, 17500, 19000, 22000, 35000]

plt.figure(figsize=(10,5))
plt.bar(meses, ventas, color='teal')
plt.title('Pronóstico de Ventas por Temporada - Ferretería')
plt.ylabel('Ventas en USD')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print(f"Mes de mayor demanda detectado: {meses[np.argmax(ventas)]}")
```

```
import matplotlib.pyplot as plt

# 1. Calcular el valor de stock por cada fila
df['Valor_Stock'] = df['Stock'] * df['Precio']

# 2. Ordenar de mayor a menor valor
df_pareto = df.sort_values(by='Valor_Stock', ascending=False).reset_index(drop=True)

# 3. Calcular porcentaje acumulado
df_pareto['Porcentaje_Acumulado'] = 100 * df_pareto['Valor_Stock'].cumsum() / df_pareto['Valor_Stock'].sum()

# 4. Graficar
plt.figure(figsize=(10,6))
plt.plot(df_pareto.index, df_pareto['Porcentaje_Acumulado'], color='blue', linewidth=2)
plt.axhline(y=80, color='r', linestyle='--', label='80% del Valor')
plt.title('Curva de Pareto - Inversión en Ferretería')
plt.xlabel('Cantidad de Productos')
plt.ylabel('% Acumulado de Inversión')
plt.legend()
plt.grid(True)
plt.show()

print("Análisis visual completado. Los productos por encima de la línea roja son tu Categoría A.")
```
